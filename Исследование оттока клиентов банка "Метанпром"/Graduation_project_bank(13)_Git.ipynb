### Проверим данные на дубликаты.</a>[К оглавлению](#up) или [к разделу](#2.2_title)

```python
# проверим все данные на наличие полных дубликатов строк
print('Дубликатов в данных:', df.duplicated().sum())
```

```python
# проверим на дубликаты пользователей

print('Дубликатов в userid:', df.duplicated('userid').sum(),
     'или {:.1%}'.format(df.duplicated('userid').sum() / df['userid'].count()))
user_dup = df[df.duplicated(subset='userid', keep=False)].sort_values(by=['userid', 'city']) # keep=False оставляет оригинал и дубликат

# настройка вывода для показа всех строк DataFrame
pd.set_option('display.max_rows', None)

# восстановление настроек вывода по умолчанию (если необходимо)
#pd.reset_option('display.max_rows')

user_dup
```

<div style="border:solid Blue 2px; padding: 10px">
    
В колонке `userid` обнаружено 73 дубликата или *0.7%* от общего количества строк в колонке. Предварительный просмотр среза по оригиналу и дубликату колонки `userid` показал, что в основном оригинал и его дубликат находятся в разных городах (Ярославль и Рыбинск, Ростова среди дубликатов нет, но это проверим.), что может свидетельствовать о наличии у одного клиента счетов в двух городах. Ниже проверим оригиналы и дубликаты на возможность нахождения в трех, двух и одном городе. 
    
Далее оригиналы и дубликаты имеют разный возраст, а самое любопытное, что оригинал и дубликат различаются гендером (М, Ж)! В остальных колонках тоже много различий между оригиналами и дубликатами, но они более-менее объяснимы.
    
В целом похоже, что оригиналы и дубликаты объединяет только колонка `userid`, в остальном - это совершенно разные записи, поэтому дубликаты предварительно не стоит удалять.

```python
# группировка по userid и подсчет уникальных городов
grouped = user_dup.groupby('userid')['city'].nunique()

# получение количества userid для трех городов
three_cities_users = grouped[grouped == 3].count()
print(f'Количество одинаковых пользователей в трех городах: {three_cities_users}')

# получение количества userid для двух городов
two_cities_users = grouped[grouped == 2].count()
print(f'Количество одинаковых пользователей в двух городах: {two_cities_users}')

# получение количества userid для одного города
one_city_users = grouped[grouped == 1].count()
print(f'Количество одинаковых пользователей в одном городе: {one_city_users}')

```

<div style="border:solid Blue 2px; padding: 10px">

**Про города в дубликатах `userid`:**
    
Пользователей, которые дублируются в разных городах нельзя считать "вредными" дубликатами. Клиент (или разные клиенты) имеет полное право иметь счет в разных отделениях/городах одного банка. Поэтому оставим данные дубликаты.  

<a id="gender_counts">
</a>

<div style="border:solid Blue 2px; padding: 10px">

**Про разный `gender` в одинаковых `userid`:**
    
    
Сделаем группировку и подсчет `userid` по `gender`, если в дубликате и оригинале `userid` одинаковый `gender`, то в графе количество будет стоять *2* ([у нас всего два значения в колонке "gender", что мы проверяли выше.)](#gender)
Если же для одного `userid` разный `gender` (М, Ж), то `userid` будет повторяться дважды, а в графе количество у каждого `userid` будет стоять количество *1*.

```python
gender_counts = user_dup.groupby(['userid', 'gender']).size().reset_index(name='count')
print(gender_counts)
```

```python
# подсчет количества дубликатов совпадающих с оригиналами по 'gender'
matching_gender_counts = gender_counts[gender_counts['count'] > 1]
print('Дубликаты совпадающие с оригиналами по "gender":', matching_gender_counts['count'].count(),
     'или {:.1%} от всех дубликатов'.format(matching_gender_counts['count'].count() / two_cities_users))
print()

# подсчет количества дубликатов различающихся с оригиналами по 'gender'
different_gender_counts = gender_counts[gender_counts['count'] == 1]
print('Дубликаты различающиеся с оригиналами по "gender":', (different_gender_counts['count'].count() / 2).astype(int),
     'или {:.1%} от всех дубликатов'.format((different_gender_counts['count'].count() / 2).astype(int) / two_cities_users))
```

<div style="border:solid Blue 2px; padding: 10px">
    
В результате сравнения `userid` дубликатов и их оригиналов по колонке `gender` выяснилось:
    
- Имеется 41 дубликат (56.2%), совпадающий с оригиналом, то есть и дубликат и оригинал имеют **одинаковый** пол (М или Ж). Это может свидетельствовать о том, что у нас один и тот же человек, родственники одинакового пола, разные люди одинакового пола.
    
- Имеется 32 дубликата (43.8%), не совпадающих с оригиналом, то есть дубликат и оригинал имеют **разный** пол (М или Ж). Это может говорить о том, что мы имеем дело с  разнополыми родственниками или разными людьми.

<div style="border:solid Blue 2px; padding: 10px">

Зайдем еще дальше и проверим наши дубликаты на совпадение по полу и возрасту. Добавим в предыдущее выражение колонку `age`. Таким образом подсчитаем *вероятных* одиночек, имеющих счета в разных городах. Вероятных, потому что это могут быть разные люди, у которых совпадает пол и возраст, объединенные одним `userid`.
    

```python
gender_age_counts = user_dup.groupby(['userid', 'gender', 'age']).size().reset_index(name='count')

alone_gender_age_counts = gender_age_counts[gender_age_counts['count'] > 1].count()

print('Вероятных одиночек со счетами в двух городах:', alone_gender_age_counts['count'],
     'или {:.1%} от всех дубликатов'.format(alone_gender_age_counts['count'] / two_cities_users))
```

<div style="border:solid Blue 2px; padding: 10px">
    
**Итоговый вывод по дубликатам:**
    
После глубокого изучения дубликатов в колонке `userid` выяснилось:
    
- Все дубликаты и оригиналы находятся в разных городах: Ярославль и Рыбинск. Дубликатов и их оригиналов представленных в одном городе не наблюдается.
    
- Под одним `userid` могут быть как мужчины, так и женщины.
    
- Под одним `userid` преимущественно находятся люди разного возраста.
    
- Есть два `userid`, у которых совпадают пол и возраст.
    
На основе вышеизложенного можно предположить, что под одним `userid` могут скрываться как родственники, так и разные люди. Данное явление может быть как умышленным (удобство для банка, специальная программа для родственников основного клиента...), так и являться ошибкой объединения данных между городами (странно, что нет Ростова).
    
Самое главное, что нет необходимости удалять дубликаты, так как за ними скрываются разные данные (дубликаты объдиняет только `userid`), что для нас приемлемо и не внесет искажения в будущие вычисления. Поэтому данные дубликаты оставляем в финальном датафрейме.

Если за 100% взять интервал 1000 - 655 = 345, то отклонение от нижнего уса до минимального значения равно 655 - 642 = 13, что составляет 13 / (345 / 100) = 3.77% от основного интервала. Это приемлемое отклонение, поэтому выбросы в колонке `score` оставляю.

## Анализ данных (определение значимых признаков отточности)</a> [К оглавлению](#up)

<div style="border:solid Blue 2px; padding: 10px">
    
Первым действие в анализе данных на этом этапе – построение матрицы корреляций по представленному датафрейму. Поскольку из 12 столбцов лишь в трех (`score`, `balance`, `salary`) представленны абсолютно количественные данные, а в остальных столбцах категоральные значения (`age` можно принять за категории, хоть их и довольно много), то для для решения данной задачи будет полезно использовать  **метод Фика ( 𝜙 k)**, поскольку он "*...последовательно работает между категориальными, порядковыми и интервальными переменными, фиксирует нелинейную зависимость и возвращается к коэффициенту корреляции Пирсона в случае двумерного нормального входного распределения.*"
    
Для сравнения построим график с корреляцией количественных данных по **методу Пирсона**, поскольку при всех преимуществах метода Фика он имеет существенный недостаток – у него нет указания направления. На всякий случай проверим направление корреляций.
    
*Но перед всем этим исключим колонку `userid` из датафрейма для рассчета корреляций - данная колонка не является признаком.*

<a id="corr1">

### Построение матрицы корреляций.</a>[К оглавлению](#up) или [К разделу](#corr1_title)


```python
#создаем датафрейм без userid
df_corr = df.drop('userid', axis=1)
df_corr.head()
```

```python
# создаем таблицу с корреляциями при помощи метода Фика
phik_overview = df_corr.phik_matrix().round(2)
phik_overview
```

```python
# создаем датафрейм с количественными переменными для Пирсона
df_quantity = df_corr[['score', 'balance', 'salary']]
df_quantity.head()
```

```python
# Вычислите корреляцию методом Пирсона
pearson_corr_matrix = df_quantity.corr(method='pearson').round(2)
pearson_corr_matrix
```

```python
# создание фигуры с подграфиками
fig, axes = plt.subplots(1, 2, figsize=(15, 6))

# Тепловая карта по методу Пирсона
sns.heatmap(pearson_corr_matrix, annot=True, cmap="coolwarm", center=0, vmin=-1, vmax=1, ax=axes[0])
axes[0].set_title("Тепловая карта по методу Пирсона")

# Тепловая карта по методу Фика
sns.heatmap(phik_overview, annot=True, cmap="coolwarm", center=0, vmin=0, vmax=1, ax=axes[1])
axes[1].set_title("Тепловая карта по методу Фика")

plt.tight_layout()
plt.show()
```

<div style="border:solid Blue 2px; padding: 10px">
    
При сравнении полученных результатов видно, что значения рассчитаные по методу Пирсона отличаются от аналогичных рассчитаных по  методу Фика. Поскольку метод Пирсона силен в количественных данных, то за основные значения корреляций `'score', 'balance', 'salary'` , будем использовать именно его. Для категоральных данных будем использовать графики посстроенные по методу Фика. Построим такой график с помощью функции *plot_correlation_matrix*  из библиотеки *phik* и сравним его для верности с первым графиком по методу Фика, который увеличим для удобства сравнения.

Напишем функцию **calculate_mean_and_median_by_churn**, которая будет выводить медианное и среднее значения данных в столбце для отточных и оставшихся клиентов.

### Расчет функции для отображения гистограммы распределения плотности вероятности оттока.</a>[К оглавлению](#up) или [к разделу](#grafic_title)


```python
'''
Функция grafic для создания гистограммы распределения плотности вероятности оттока. В качестве аргументов принимает:
датафрейм, колонку для оси Х, название графика (если название не задано, отображается название по-умолчанию),
название оси Х (если название оси Х, то отображается название колонки из второго аргумента).
Для определенных колонок из списка, который задается в теле функции использует palette = 'muted',
для остальных palette = 'flag'. 

'''

def grafic(dataframe, column, title=None, xlabel=None):
    plt.figure(figsize=(10, 6))
    
    # условное выражение для выбора палитры в зависимости от значения column
    palette = 'muted' if column in ['products', 'equity', 'score', 'last_activity', 'gender'] else 'flag'
    
    sns.histplot(data=dataframe, hue='churn', x=column, stat='density', common_norm=False, palette=palette)
    
    # если указано название графика, то установите его
    if title:
        plt.title(title)
    else:
        plt.title(f'Гистограмма распределения плотности вероятности оттока по {column}')
    
    # если указана подпись оси Х, то установите его
    if xlabel:
        plt.xlabel(xlabel)
    else:
        plt.xlabel(column)
     
    plt.ylabel('Плотность распределения вероятности')
    plt.tight_layout()
    plt.show()
```

<a id="def">

### Расчет функции для вычисления *median* и *mean* для ушедших и оставшихся клиентов.</a>[К оглавлению](#up) или [к разделу](#def_title)

```python
'''
Функция calculate_mean_and_median_by_churn() принимает в качестве аргумента датафрейм, а выводит на печать
медианное и среднее значения столбца при churn == 1 и churn == 0, для чего в начале функции создаются два 
датафрейма с соответствующими условиями. Принимает только столбцы с числовыми значениями,
что проверяется в теле функции выражением: if dataframe[column].dtype != 'object':
Выводимые значения округляются до целого цисла.

'''

def calculate_mean_and_median_by_churn(dataframe):
    # создадим два датафрейма с churn == 1 и churn == 0
    churn_1 = dataframe[dataframe['churn'] == 1]
    churn_0 = dataframe[dataframe['churn'] == 0]

    # вычислим среднее и медианное значение для каждого столбца
    for column in dataframe.columns:
        if column != 'churn':
            if dataframe[column].dtype != 'object':  # проверяем тип данных столбца                
                median_churn_1 = round(churn_1[column].median())                
                median_churn_0 = round(churn_0[column].median())
                mean_churn_1 =  round(churn_1[column].mean())
                mean_churn_0 =  round(churn_0[column].mean())

                print(f'Медиана отточных клиентов в колонке {column} равна: {median_churn_1}')
                print(f'Медиана оставшихся клиентов в колонке {column} равна: {median_churn_0}')
                print(f'Среднее отточных клиентов в колонке {column} равно: {mean_churn_1}')
                print(f'Среднее оставшихся клиентов в колонке {column} равно: {mean_churn_0}')
```

## Создание и проверка статистических гипотез.</a>[К оглавлению](#up) или [к разделу](#hypotheses_title)



Поскольку будут проверяться гипотезы по оттоку клиентов, то создадим два датафрейма с оставшимися(`churn_0`) и ушедшими(`churn_1`) клиентами банка. Уровень статистической значимости (alpha) зададим равным 5%. также создадим две функции для проведения тестов.

```python
churn_0 = df.query('churn == 0')

churn_1 = df.query('churn == 1')

alpha = 0.05
```

```python
'''
Функция для проведения теста о равенстве среднего двух генеральных совокупностей. В качестве параметров принимаются два
датафрейма с данными, прараметр "equal_var", принимающий "True", "False" и параметр "alternative",
который может принимать три значения: "two-sided"(по умолчанию), "less", "greater". Значение alpha задается вне тела функции.

'''

def ttest_ind(df_0, df_1, equal_var=None, alternative='two-sided'):
    
    results = st.ttest_ind(df_0, df_1, alternative=alternative) 
    
    print('p-значение:', results.pvalue)
    
    if results.pvalue < alpha:
        print('Отвергаем нулевую гипотезу')
    else:
        print('Не получилось отвергнуть нулевую гипотезу')
```

```python
'''
Функция для проведения теста Уилкоксона-Манна-Уитни для сравнения медиан двух генеральных совокупностей.
В качестве параметров принимаются два датафрейма с данными и параметр "alternative",
который может принимать три значения: "two-sided"(по умолчанию), "less", "greater". Значение alpha задается вне тела функции.

'''

def test_mannwhitneyu(df_0, df_1, alternative='two-sided'):

    results = mannwhitneyu(df_0, df_1, alternative=alternative)

    print('p-значение:', results.pvalue)

    if results.pvalue < alpha:
        print('Отвергаем нулевую гипотезу')
    else:
        print('Не получилось отвергнуть нулевую гипотезу')
```

## Создание и отбор сегментов по признакам отточности.</a> [К оглавлению](#up) или [к разделу](#segmentation_title)

```python
'''
Функция churn_df на вход получает необходимый датафрейм, исследуемый столбец и условие для среза методом
query в одинарных кавычках. На выходе печатает количество значений в отфильтрованном столбце, печатает результат
групировки по оттоку 'churn' в отфильтрованном стобце  и отношение (долю) отточных к оставшимся в процентах в отфильтрованном
столбце. Также выводит процент отточных от общего числа в отфильтрованном столбце.

'''    
def churn_df(dataframe, column, condition_query=None):
    if condition_query is not None:
        filtered_df = dataframe.query(condition_query)
        print('Общее количество клиентов в столбце', column, 'равно:', filtered_df[column].count())
        print()
        grouped_filtered_df = filtered_df.groupby('churn')['userid'].count().reset_index(name='count')# +'_'+ column)
        print(grouped_filtered_df)
        print()
        
        # расчет отношения churn[1] к churn[0]
        churn_1_count = grouped_filtered_df.loc[grouped_filtered_df['churn'] == 1, 'count'].values[0]
        churn_0_count = grouped_filtered_df.loc[grouped_filtered_df['churn'] == 0, 'count'].values[0]
        churn_ratio = churn_1_count / churn_0_count
        
        
        formatted_ratio = "{:.0%}".format(churn_ratio)
        print(f'Отношение в столбце "{column}" ушедших к оставшимся: {formatted_ratio}')
        print()
        
        total_ratio = churn_1_count / filtered_df[column].count()
        formatted_total_ratio = "{:.0%}".format(total_ratio)
        print(f'Отношение в столбце "{column}" ушедших к общему количеству в колонке: {formatted_total_ratio}')
        
    
    else:
        print("Задайте в параметр функции condition_query = 'тут ваше условие'")
    

# Пример использования с query-срезом:
#churn_df(df, 'score', 'score >=821 & score <=910')
```

Используя 2-3 признака отточных клиентов сформируем финальные сегменты.

Выделим финальные сегменты, для чего напишем предварительно функцию, которая будет по полученному срезу выдавать количество клиентов и отношение ушедших к оставшимся. Количество должно укладываться в диапазон 500 - 2000 клиентов, а процент ушедших к оставшимся должен быть чем выше, тем лучше. Прогоним каждый сегмент через все остальные, предварительно оставим только те, где количество выше 500 клиентов.

```python
'''
Функция df_query получает на вход датафрейм и условие для среза методом query в одинарных кавычках. На выходе выводит 
количество клиентов в срезе, результат группировки среза по столбцу churn(по желанию, по умолчанию закомментировано),
процентное отношение ушедших клиентов к оставшимся (churn[1] / churn[0] * 100). Имеется параметр min_person, в котором
задается минимальный порог количества в срезе по умолчанию равный 500. В зависимости от параметра и результата в первом
принте выводится информационное сообщение о прохождении минимального порога "...Ура! Подходит!" или о недостаточном
количестве значений в срезе "...Маловато будет!". Также есть параметр max_percent, в котором задается максимальный
процент ушедших (по умолчанию 50). В случае, если полученный результат больше или равен этому значению, выводится 
информационное сообщение "...Побег из Шоушенка!", сигнализирующее об оттоке 2/3 клиентов. В конце выводится процент
отточных клиентов от общего их количества в получившемся срезе.

'''

def df_query(dataframe, condition_query=None, min_person=500, max_percent=50):
    if condition_query is not None:
        filtered_df = dataframe.query(condition_query)
        if filtered_df['userid'].count() >= min_person:
            print('Общее количество клиентов в срезе равно:', filtered_df['userid'].count(), '...Ура! Подходит!')
        else:
            print('Общее количество клиентов в срезе равно:', filtered_df['userid'].count(), '...Маловато будет!')
        #print()
        grouped_filtered_df = filtered_df.groupby('churn')['userid'].count().reset_index(name='count')# +'_'+ column)
        #print(grouped_filtered_df)
        #print()
        
        # расчет отношения churn[1] к churn[0]
        churn_1_count = grouped_filtered_df.loc[grouped_filtered_df['churn'] == 1, 'count'].values[0]
        churn_0_count = grouped_filtered_df.loc[grouped_filtered_df['churn'] == 0, 'count'].values[0]
        churn_ratio = churn_1_count / churn_0_count * 100
        if churn_ratio >= max_percent:
            print(f'Отношение ушедших к оставшимся: {churn_ratio:.0f}%  ...Побег из Шоушенка!')
        else:
            print(f'Отношение ушедших к оставшимся: {churn_ratio:.0f}%')
        
        if filtered_df['userid'].count() >= min_person and churn_ratio >= max_percent:
            print('На такое больше нельзя закрывать глаза!')
            
        # расчет отношения churn[1] к filtered_df['userid'].count()   
        churn_total_ratio = (churn_1_count / filtered_df['userid'].count()) * 100
        print(f'Процент ушедших от общего количества в срезе: {churn_total_ratio:.0f}%')
               
    else:
        print("Задайте в параметр функции condition_query = 'тут ваше условие'")
    
```
